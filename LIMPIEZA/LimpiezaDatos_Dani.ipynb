```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/GorgeLop07/HACKATON---InnovaLab/refs/heads/main/Dataset_Limpio_para_Participantes.csv")

df.head()
```

```python
df.info()
```

```python
calidad_map = {"Baja": 0, "Media": 1, "Alta": 2}
df["calidad_auto_num"] = df["calidad_auto"].map(calidad_map)
```

```python
correlaciones = df.corr(numeric_only=True)
print(correlaciones["calidad_auto_num"].sort_values(ascending=False))
```

```python
df_limpio = df.drop(columns=[
    "name", "calidad_auto",
    "fuel", "seller_type", "km_driven", "tipo_carroceria",
    "owner", "combustible_estimado_l"
])
```

```python
df_limpio.to_csv("Dataset_ML_Limpio_V1.csv", index=False)
```
