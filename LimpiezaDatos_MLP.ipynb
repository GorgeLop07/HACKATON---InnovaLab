# **Limpieza de Datos**
Aqui podemos observar e ir poniendo como vamos a ir limpiando el excel utilizando herramientas como Pandas y skLearn:

En este archivo estaremos utilizando las siguientes herramientas para la limpieza de dato:

1 - One-Hot Encoding

2 - Modificacion de valores a valores numericos y asegurarse que no existan valores N/A

3 - Normalizamos los valores con StandardScaler de esta forma los valores puedan pasar mejor por nuestra arquitectura

**Esta base de datos generada al final se utilizara en una arquitectura de Deep Learning llamada MLP (Multilayer Perceptron)**


```python
import pandas as pd

#Importamos nuestra base de datos limpia
df = pd.read_excel('/content/Dataset_Limpio_para_Participantes.xlsx')

#borro la columna de nombre
df = df.drop('name', axis=1)

df
```

# **Generacion de Dummies dentro del Excel**

Explicacion:
Por generar bien (Pero en resumen se hacen dummies)

**Seccion 1:**
En esta seccion establecemos las columnas a las que les queremos generar dummies para despues cambiarles los nombres a valores mas entendibles

```python
#Declaracion de nombres a los que se les aplicara el Dummie
columas_One_Hot_Encoding = ['fuel', 'seller_type', 'tipo_carroceria']
columnas_binarias = ['trasmission']
columna_Target = 'calidad_auto'

#Se realiza el one-Hot-Encoding para asi generar los Dummies
df_Dummies = pd.get_dummies(df, columns=columas_One_Hot_Encoding, drop_first=False)

#Utilizo el metodo de Rename para camibiar los nombres a algo mas entendible
df_renombrado = df_Dummies.rename(columns={
    'fuel_0': 'Petrol',
    'fuel_1': 'Diesel',
    'fuel_2': 'CNG',
    'fuel_3': 'LPG',
    'fuel_4': 'Electric',
    'seller_type_0': 'Individual',
    'seller_type_1': 'Dealer',
    'seller_type_2': 'Trustmark Dealer',
    'tipo_carroceria_0': 'Sedan',
    'tipo_carroceria_1': 'Hatchback',
    'tipo_carroceria_2': 'SUV',
    'tipo_carroceria_3': 'Pickup',
    'tipo_carroceria_4': 'Deportivo',
})

# Comparacion entre nuevos headers de datos
display(df_renombrado.head())
display(df_Dummies)
```

**Seccion 2:** Aqui lo que hicimos fue el modificar todos los valores booleanos o de string a valores numericos tomando la siguiente nomenclatura:

**Dentro de nuestros Dummies:**

True - 1

False - 0

**Dentro de calidad o valor predicho:**

"Alta" - 3


"Media" - 2


"Baja" - 1

Tambien lo que se realiza es un algoritmo para robustecer la limpieza de los datos, asegurandose que solo existan los valores deseados dentro de las columas

```python
# Selecciona las columnas con tipo de dato booleano y cambiarlos a booleanos
columnas_Boleanas = df_renombrado.select_dtypes(include='bool').columns
df_renombrado[columnas_Boleanas] = df_renombrado[columnas_Boleanas].astype(int)

#Esta siguiente parte solo sirve para asegurarse que solo esten los valores que esperamos tener para asi evitar errores
Valores_de_calidad = df_renombrado['calidad_auto'].unique()

Valores_Esperados = ['Alta', 'Media', 'Baja']

valores_inesperados = [valor for valor in Valores_de_calidad if valor not in Valores_Esperados]

if valores_inesperados:
    print("Se encontraron valores inesperados en la columna 'calidad_auto':", valores_inesperados)
    print("Puedes considerar eliminar las filas con estos valores o reemplazarlos por un valor nulo o por defecto.")
else:
    print("La columna 'calidad_auto' solo contiene los valores esperados: 'Alta', 'Media', 'Baja'.")


# Define el mapeo de los valores de texto a numéricos
mapeo_calidad = {
    'Alta': 2,
    'Media': 1,
    'Baja': 0
}

# Aplicar el cambio de datso
df_renombrado['calidad_auto'] = df_renombrado['calidad_auto'].map(mapeo_calidad)

display(df_renombrado.head())
```

```python
#Solo por estetica movemos el valor a predecir a la ultima columna
columnas = df_renombrado.columns.tolist()
columnas.remove('calidad_auto')
columnas.append('calidad_auto')
df_renombrado = df_renombrado[columnas]



display(df_renombrado.head())
```

# **Aplicar Normalizacion de Datos**

En esta seccion estaremos Standarizando los valores para que no esten en escalas tan distintas, de esta forma podemos evitar que el modelo de Deep Learning le de importancia mayor a valores mas grandes

```python
from sklearn.preprocessing import StandardScaler

# Quitar las columnas que no queremos normalizar
columnas_a_normalizar = df_renombrado.columns.tolist()
columnas_a_normalizar.remove('calidad_auto')

# Identifica las columnas dummy (las que terminan con 0, 1, 2, etc. de los pasos anteriores)
# Una forma de hacerlo es identificando las columnas que fueron creadas por get_dummies
# Basándonos en los nombres renombrados y los originales, podemos inferir cuáles son.
# Las columnas dummy renombradas son: 'Petrol', 'Diesel', 'CNG', 'LPG', 'Electric',
# 'Individual', 'Dealer', 'Trustmark Dealer', 'Sedan', 'Hatchback', 'SUV', 'Pickup',
# 'Deportivo', y la columna original no renombrada 'tipo_carroceria_5'

columnas_dummy_renombradas = [
    'Petrol', 'Diesel', 'CNG', 'LPG', 'Electric',
    'Individual', 'Dealer', 'Trustmark Dealer', 'Sedan', 'Hatchback',
    'SUV', 'Pickup', 'Deportivo', 'tipo_carroceria_5', 'transmission', 'owner']


# Se filtra las columnas Dummy y las columans con valores pequenos
columnas_para_scalar = [col for col in columnas_a_normalizar if col not in columnas_dummy_renombradas]

# Aqui solo se selecciona las columans a escalar y nos podemos asegurar que solo sean valores numericos
columnas_para_scalar = df_renombrado[columnas_para_scalar].select_dtypes(include=['int64', 'float64']).columns

print("Columnas que serán normalizadas:", columnas_para_scalar.tolist())

#Se declara el objeto que nos permitira genera la normalizacion para despues aplicarlo a las columanas que queremos
scaler = StandardScaler()
df_renombrado[columnas_para_scalar] = scaler.fit_transform(df_renombrado[columnas_para_scalar])
display(df_renombrado.head())
```
